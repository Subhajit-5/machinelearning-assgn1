4 Volume of a sphere with radius 12

```python
import math
def volumeofsphere(r):
    v=(4/3)*math.pi*(r*r*r)
    return(v)

radius=12
volume=volumeofsphere(radius)
print(volume)
```

6 print pattern

```python
for i in range(1,9):
    k=1
    lt=0
    ut=i
    if i>5:
        k=-1
        lt=9-i
        ut=0
    for j in range(lt,ut,k):
        print("*",end="")
    print('\n')
        
```
